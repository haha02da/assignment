```python
s = "abab"
p = "ab"
```

```python

def findAnagrams(s: str, p: str):
    short=list(p)
    long=list(s)
    answer=[]
    for i in range(len(long)-len(short)+1):
        count={}
        test=long[i:i+len(short)]
        for j in test:
            count[j]=count.get(j,0)
        count_ans={}
        for k in short:
            count_ans[k]=count.get(k,0)
        if count==count_ans:
            answer.append(i)
    print(answer)
```

```python
findAnagrams(s,p)
```
